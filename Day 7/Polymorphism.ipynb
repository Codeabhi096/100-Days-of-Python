# üß© 1. Python ‚Äì **Polymorphism**

### üîπ What is Polymorphism?

The term **Polymorphism** means **‚Äúmany forms‚Äù**.
In Object-Oriented Programming (OOP), it allows **the same function or method name** to behave **differently depending on the object or data type**.

---

### üîπ Real-Life Analogy

üëâ Example:
The function `drive()` behaves differently for:

* a `Car` object (drives on road),
* a `Boat` object (drives on water),
* an `Airplane` object (flies in air).

Same **interface**, but **different behaviors** ‚Äî that‚Äôs polymorphism.

---

### üîπ Types of Polymorphism in Python

| Type                      | Description                                                    | Example                        |
| ------------------------- | -------------------------------------------------------------- | ------------------------------ |
| **Compile-time (Static)** | Decided at compile-time (Python doesn‚Äôt support this directly) | Method overloading             |
| **Runtime (Dynamic)**     | Decided at runtime based on object type                        | Method overriding, duck typing |

---

## üß† Type 1: Polymorphism with Functions and Objects

```python
class Dog:
    def sound(self):
        print("Barks üê∂")

class Cat:
    def sound(self):
        print("Meows üê±")

# Function using polymorphism
def make_sound(animal):
    animal.sound()  # same function, different behavior

d = Dog()
c = Cat()

make_sound(d)
make_sound(c)
```

üß† **Output**

```
Barks üê∂
Meows üê±
```

‚úîÔ∏è **Explanation:**
`make_sound()` works for any object that has a `sound()` method ‚Üí this is called **duck typing** ("if it quacks like a duck...").

---

## üß† Type 2: Polymorphism with Class Methods

```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        print("Area of Circle = œÄr¬≤")

class Square(Shape):
    def area(self):
        print("Area of Square = side¬≤")

shapes = [Circle(), Square()]

for s in shapes:
    s.area()
```

üß† **Output**

```
Area of Circle = œÄr¬≤
Area of Square = side¬≤
```

‚úîÔ∏è **Explanation:**
Each subclass defines its own version of the method `area()`.

---

## üß† Type 3: Polymorphism with Inheritance (Dynamic Polymorphism)

It happens when **a child class redefines a parent method**, and the **method to be executed is chosen at runtime**.

```python
class Bird:
    def fly(self):
        print("Most birds can fly üïäÔ∏è")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly üêß")

b1 = Bird()
b2 = Penguin()

b1.fly()  # calls Bird‚Äôs method
b2.fly()  # calls Penguin‚Äôs method
```

üß† **Output**

```
Most birds can fly üïäÔ∏è
Penguins cannot fly üêß
```

‚úîÔ∏è **Explanation:**
Even though both have the same `fly()` method, behavior differs depending on the object ‚Üí **runtime polymorphism**.

# üß© 2. Python ‚Äì **Method Overriding**

### üîπ What is Method Overriding?

When a **child class defines a method with the same name as a parent class**, the **child‚Äôs version overrides** the parent‚Äôs version.

This is **runtime polymorphism** (decided at execution time).

---

### üîπ Syntax Example

```python
class Parent:
    def show(self):
        print("This is Parent class method")

class Child(Parent):
    def show(self):  # overriding
        print("This is Child class method")

obj = Child()
obj.show()
```

üß† **Output**

```
This is Child class method
```

‚úîÔ∏è **Explanation:**
The `show()` method in the child class **replaces** the one from the parent.

---

### üîπ Using `super()` to Access the Parent Method

```python
class Parent:
    def display(self):
        print("Parent display() called")

class Child(Parent):
    def display(self):
        super().display()  # call parent method
        print("Child display() called")

obj = Child()
obj.display()
```

üß† **Output**

```
Parent display() called
Child display() called
```

‚úîÔ∏è **Explanation:**
`super()` lets you **reuse** the parent‚Äôs method before or after adding new behavior.

---

### üîπ Internal Working (Memory + MRO)

When you call a method on an object:

1. Python checks the **instance‚Äôs class**.
2. If not found, it checks **parent classes** (left-to-right) according to **Method Resolution Order (MRO)**.
3. First match found ‚Üí executed.

Check MRO:

```python
print(Child.mro())
```

# üß© 3. Python ‚Äì **Method Overloading**

### üîπ What is Method Overloading?

**Method overloading** means defining **multiple methods with the same name but different parameters**.

‚û°Ô∏è In **Python**, true overloading like Java/C++ is **not supported**,
because Python **does not check parameter types or count** during function definition.

üëâ Instead, **the latest defined method overrides previous ones**.

---

### üîπ Example (What actually happens)

```python
class Demo:
    def add(self, a, b):
        print(a + b)

    def add(self, a, b, c):  # overrides previous add()
        print(a + b + c)

d = Demo()
# d.add(2, 3)     ‚ùå Error: missing argument
d.add(2, 3, 4)    # ‚úÖ works
```

üß† **Output**

```
9
```

‚úîÔ∏è **Explanation:**
The second definition of `add()` overwrote the first ‚Üí Python keeps **only the last one**.

---

### üîπ Simulating Method Overloading (using Default Arguments)

```python
class Demo:
    def add(self, a=None, b=None, c=None):
        if a is not None and b is not None and c is not None:
            print(a + b + c)
        elif a is not None and b is not None:
            print(a + b)
        else:
            print("Provide at least two numbers!")

d = Demo()
d.add(2, 3)
d.add(2, 3, 4)
d.add(2)
```

üß† **Output**

```
5
9
Provide at least two numbers!
```

‚úîÔ∏è **Explanation:**
We manually check arguments to simulate different behaviors.

---

### üîπ Simulating Overloading with `*args`

```python
class MathOp:
    def add(self, *args):
        print("Sum =", sum(args))

m = MathOp()
m.add(5, 10)
m.add(5, 10, 15, 20)
```

üß† **Output**

```
Sum = 15
Sum = 50
```

‚úîÔ∏è **Explanation:**
`*args` accepts variable numbers of arguments, simulating overloading.

# ‚öôÔ∏è Internal Working in Memory (Simplified)

When an object is created:

1. Python allocates memory for its **instance dictionary (`__dict__`)**.
2. When calling a method, Python checks:

   * **Object‚Äôs class**
   * Then **Parent classes** using **MRO**
3. Whichever method name is found first ‚Üí executed.

Example:

```python
print(Demo.__dict__)
```

This shows how methods are stored and how the latest one replaces the previous (hence no true overloading).
