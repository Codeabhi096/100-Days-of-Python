# üß© 1. Python ‚Äì **Constructors**

### üîπ What is a Constructor?

A **constructor** is a special method that runs **automatically when an object is created**.
It is mainly used to **initialize the attributes (variables)** of the class.

In Python, constructors are defined using:

```python
def __init__(self, ...):
```

---

### üîπ Types of Constructors in Python

#### 1Ô∏è‚É£ Default Constructor

A constructor with **no parameters** (other than `self`).
Used when all attributes have default values.

```python
class Student:
    def __init__(self):
        self.name = "Unknown"
        self.age = 0

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Object creation
s1 = Student()      # default constructor called automatically
s1.show()
```

üß† **Output**

```
Name: Unknown, Age: 0
```

---

#### 2Ô∏è‚É£ Parameterized Constructor

Constructor that **takes parameters** to initialize object attributes dynamically.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Object creation
s1 = Student("John", 21)
s2 = Student("Emma", 19)

s1.show()
s2.show()
```

üß† **Output**

```
Name: John, Age: 21
Name: Emma, Age: 19
```

---

#### 3Ô∏è‚É£ Constructor with Default Arguments

When parameters have default values.

```python
class Student:
    def __init__(self, name="Guest", age=18):
        self.name = name
        self.age = age

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

s1 = Student()
s2 = Student("Alice", 22)

s1.show()
s2.show()
```

üß† **Output**

```
Name: Guest, Age: 18
Name: Alice, Age: 22
```

---

#### 4Ô∏è‚É£ Dynamic / Overloaded Constructor (Simulated)

Python doesn‚Äôt support multiple constructors (like Java/C++).
But we can simulate using `*args` or `**kwargs`.

```python
class Student:
    def __init__(self, *args):
        if len(args) == 0:
            self.name = "Unknown"
            self.age = 0
        elif len(args) == 1:
            self.name = args[0]
            self.age = 0
        else:
            self.name = args[0]
            self.age = args[1]

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

s1 = Student()
s2 = Student("Maya")
s3 = Student("Raj", 20)

s1.show()
s2.show()
s3.show()
```

üß† **Output**

```
Name: Unknown, Age: 0
Name: Maya, Age: 0
Name: Raj, Age: 20
```

# üß© 2. Python ‚Äì **Access Modifiers**

### üîπ What are Access Modifiers?

Access modifiers define **visibility** or **scope** of variables and methods.
Python uses **naming conventions** rather than strict enforcement (like Java or C++).

| Modifier      | Syntax        | Meaning                                                  |
| ------------- | ------------- | -------------------------------------------------------- |
| **Public**    | No underscore | Accessible from anywhere                                 |
| **Protected** | `_variable`   | Accessible within class and subclasses (convention only) |
| **Private**   | `__variable`  | Name mangled (hidden) ‚Äì accessible only within class     |

---

### üîπ Public Members Example

```python
class Student:
    def __init__(self, name, age):
        self.name = name      # public attribute
        self.age = age

s1 = Student("John", 20)
print(s1.name)  # Accessible
print(s1.age)
```

üß† Output:

```
John
20
```

---

### üîπ Protected Members Example (`_single_underscore`)

It‚Äôs **just a convention** ‚Äî not strictly private, but signals ‚Äúfor internal use‚Äù.

```python
class Student:
    def __init__(self):
        self._school = "ABC Public School"  # protected attribute

class Kid(Student):
    def show_school(self):
        print("School:", self._school)  # accessible in subclass

k = Kid()
k.show_school()
print(k._school)  # still accessible, but discouraged
```

üß† Output:

```
School: ABC Public School
ABC Public School
```

---

### üîπ Private Members Example (`__double_underscore`)

Python **name-mangles** private members:
`__var` becomes `_ClassName__var` internally.

```python
class Student:
    def __init__(self):
        self.__grade = "A+"  # private

    def show(self):
        print("Grade:", self.__grade)

s1 = Student()
s1.show()

# print(s1.__grade)        ‚ùå Error: AttributeError
print(s1._Student__grade)  # ‚úÖ Access via name-mangling
```

üß† Output:

```
Grade: A+
A+
```

# üß© 3. Python ‚Äì **Inheritance**

### üîπ What is Inheritance?

**Inheritance** allows a new class (child) to **reuse attributes and methods** of another class (parent).
It promotes **code reuse** and **hierarchical relationships**.

---

### üîπ Types of Inheritance in Python

#### 1Ô∏è‚É£ **Single Inheritance**

Child inherits from one parent class.

```python
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def hello(self):
        print("Hi from Child")

obj = Child()
obj.greet()
obj.hello()
```

üß† Output:

```
Hello from Parent
Hi from Child
```

---

#### 2Ô∏è‚É£ **Multilevel Inheritance**

Child inherits from a parent, and that parent inherits from another.

```python
class Grandparent:
    def feature1(self):
        print("Grandparent feature")

class Parent(Grandparent):
    def feature2(self):
        print("Parent feature")

class Child(Parent):
    def feature3(self):
        print("Child feature")

c = Child()
c.feature1()
c.feature2()
c.feature3()
```

üß† Output:

```
Grandparent feature
Parent feature
Child feature
```

---

#### 3Ô∏è‚É£ **Hierarchical Inheritance**

Multiple child classes share the same parent.

```python
class Parent:
    def show(self):
        print("Parent feature")

class Child1(Parent):
    def c1(self):
        print("Child1 feature")

class Child2(Parent):
    def c2(self):
        print("Child2 feature")

obj1 = Child1()
obj2 = Child2()

obj1.show()
obj2.show()
```

üß† Output:

```
Parent feature
Parent feature
```

---

#### 4Ô∏è‚É£ **Hybrid Inheritance (Mixed)**

Combination of multiple inheritance types.
(Handled via **MRO ‚Äì Method Resolution Order**)

```python
class A:
    def show(self): print("A")
class B(A):
    def show(self): print("B")
class C(A):
    def show(self): print("C")
class D(B, C):
    pass

d = D()
d.show()
print(D.mro())
```

üß† Output:

```
B
[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

# üß© 4. Python ‚Äì **Multiple Inheritance**

### üîπ What is Multiple Inheritance?

When a **child class inherits from more than one parent class**.

```python
class Father:
    def car(self):
        print("Father's Car")

class Mother:
    def house(self):
        print("Mother's House")

class Child(Father, Mother):
    def own(self):
        print("Child‚Äôs Own Bike")

c = Child()
c.car()
c.house()
c.own()
```

üß† Output:

```
Father's Car
Mother's House
Child‚Äôs Own Bike
```

---

### üîπ When Methods Have the Same Name (Diamond Problem)

Python uses **MRO (Method Resolution Order)** to decide **which method to call first**.

```python
class A:
    def show(self): print("A")
class B(A):
    def show(self): print("B")
class C(A):
    def show(self): print("C")
class D(B, C):
    pass

d = D()
d.show()
print(D.mro())  # check MRO order
```

üß† Output:

```
B
[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

‚úÖ **MRO** follows the **C3 linearization algorithm** to prevent ambiguity.
